## 5) Atividade 3 — KPIs do negócio

Considere que você é analista de um varejista e precisa apresentar indicadores para a gestão.

Calcule os seguintes KPIs:

1. **Receita total**
2. **Lucro total**
3. **Ticket médio** (`Receita média por registro`)
4. **Quantidade total vendida**

Depois, escreva uma breve interpretação:
- O desempenho geral parece positivo?
- Receita e lucro caminham juntos?

```python
receita_total = df['Receita'].sum()
lucro_total = df['Lucro'].sum()
ticket_medio = df['Receita'].mean()
quantidade_total = df['Quantidade'].sum()

print("Receita total:", receita_total)
print("Lucro total:", lucro_total)
print("Ticket médio:", ticket_medio)
print("Quantidade total vendida:", quantidade_total)
```

Resposta

O desempenho geral é positivo, pois há geração de receita e lucro.
Receita e lucro caminham juntos, já que o lucro é derivado das vendas — quando a receita aumenta, o lucro tende a acompanhar, desde que os custos estejam controlados

## 6) Atividade 4 — Respondendo perguntas do gestor com dados

Use o dataset para responder:

1. **Qual canal gera mais receita?**
2. **Qual produto gera mais lucro?**
3. **Quais UFs concentram maior receita?**

Para cada resposta:
- mostre o código;
- apresente o resultado;
- escreva uma interpretação curta.

```python
canal_receita = df.groupby('Canal_Venda')['Receita'].sum().sort_values(ascending=False)
print(canal_receita)

produto_lucro = df.groupby('Produto')['Lucro'].sum().sort_values(ascending=False)
print(produto_lucro)

uf_receita = df.groupby('UF')['Receita'].sum().sort_values(ascending=False)
print(uf_receita)
```

Resposta

O E-commerce é o principal canal de vendas, indicando forte presença digital.
O produto Tênis é o mais lucrativo, podendo ser priorizado em estratégias.
As UFs SP, RJ e MG concentram a maior receita, mostrando que o Sudeste é o principal mercado consumidor.

## Desafio extra (opcional)

Faça uma pequena limpeza no dataset:

- trate valores nulos;
- padronize os nomes dos canais de venda;
- converta `Data_Venda` para formato de data.

Depois, compare os resultados **antes e depois da limpeza**.

```python
df_limpo = df.copy()

df_limpo['Canal_Venda'] = df_limpo['Canal_Venda'].str.lower()
df_limpo['Canal_Venda'] = df_limpo['Canal_Venda'].str.replace('-', '').str.strip()

df_limpo['Data_Venda'] = pd.to_datetime(df_limpo['Data_Venda'])

df_limpo = df_limpo.fillna({
    'Canal_Venda': df_limpo['Canal_Venda'].mode()[0],
    'UF': df_limpo['UF'].mode()[0],
    'Preco_Unitario': df_limpo['Preco_Unitario'].mean()
})

print("ANTES:")
print(df.isnull().sum())

print("\nDEPOIS:")
print(df_limpo.isnull().sum())
```
